```python
import pandas as pd
import numpy as np
import csv
import json
import pickle
import tqdm
```

```python
#企業リストを取得
company_list = pd.read_csv("../data/new2/df_achieve/company_list.csv",index_col=0)
```

```python
company_list=company_list.drop(14350)
```

```python
company_list
```

```python
com_dic={}
#予測のない企業
for i,company in enumerate(company_list.index):
    df_path= '../data/new2/df_achieve/this/{}.csv'.format(company)
    df=pd.read_csv(df_path,index_col=0)   
    com_dic[company]={"name":"","date":[]}
    com_dic[company]["name"]=company_list.loc[company]["name"]
    com_dic[company]["date"]=[]
    for i in range(0,len(df)):
        if not "訂正" in df["title"][i] and  not "四半期" in df["title"][i] and df["achieve"][i]=="None":
            com_dic[company]["date"].append(df.index[i])
```

```python
com_dic
```

```python
cnt=0
for company in com_dic.keys():
    if com_dic[company]["date"]!=[]:
        if com_dic[company]["date"][0][:4]!="2020":
            print(company,com_dic[company])
            cnt+=1
```
