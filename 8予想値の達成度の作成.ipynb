```python
def FJ_score(dis_number):
    if txt_df["word_cnt"][dis_number]!=0:
        f_score=(txt_df["f_pos"][dis_number]-txt_df["f_neg"][dis_number])/txt_df["word_cnt"][dis_number]
        j_score=(txt_df["j_pos"][dis_number]-txt_df["j_neg"][dis_number])/txt_df["word_cnt"][dis_number]
    else:
        f_score=0
        j_score=0
    return f_score,j_score
```

```python
for company in tqdm.tqdm(company_list.index):
    path = '../data/new2/df_achieve/this/{}.csv'.format(company,company)
    df = pd.read_csv(path,index_col=0)
    
    achieve=[]
    achieve_list=list(df["achieve"])
    uriage_list=list(df["uriage"])
    pred_list=list(df["pred"])
    tuki_pred_list=list(df["tuki_pred"])
    title_list=list(df["title"])
    
    for i,pred in enumerate(pred_list):
        try:
            if pred=="None":
                achieve.append("None")

            elif i!=0 and achieve[-1]=="None" and "四半期" in title_list[i]:
                achieve.append("None")

            elif uriage_list[i]=="None":
                achieve.append("None")
                
                #期末決算の時
            elif i!=0 and not "四半期" in title_list[i]:
                achieve.append((float(uriage_list[i]) - float(tuki_pred_list[i-1])) / float(tuki_pred_list[i-1]))
                #四半期決算
            else:
                achieve.append((float(uriage_list[i]) - float(tuki_pred_list[i])) / float(tuki_pred_list[i]))
        except:
            achieve.append("None")
    df["achieve"]=achieve
    df.to_csv(path)
```
