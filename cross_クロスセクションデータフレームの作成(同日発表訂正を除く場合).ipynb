```python
def create_df(df):
    df.index=df["date"]
    col= list(df.columns)
    col.remove("number.1")
    col.remove("title")
    col.remove("date")
    df=df.reindex(columns=col)
    #df=df.rename(columns={"number.1": 'pdf_code'})
    return(df,col)
```

```python
def code_create_df(df):
    df.index=df["date"]
    col= list(df.columns)
    #col.remove("number.1")
    #col.remove("title")
    col.remove("date")
    df=df.reindex(columns=col)
    df=df.rename(columns={"number.1": 'pdf_code'})
    col= list(df.columns)
    return(df,col)
```

```python
import re
def normalize_text(text1):
    text1=re.sub("０", "0", text1)
    text1=re.sub("１", "1", text1)
    text1=re.sub("２", "2", text1)
    text1=re.sub("３", "3", text1)
    text1=re.sub("４", "4", text1)
    text1=re.sub("５", "5", text1)
    text1=re.sub("６", "6", text1)
    text1=re.sub("７", "7", text1)
    text1=re.sub("８", "8", text1)
    text1=re.sub("９", "9", text1)
    return(text1)
```
