ROA（％）＝当期純利益÷総資産×100

```python
import time
#はじめのlap=timeは−１
#start_time = time.time()
#lap_time = -1
#for
#    lap_time=progress_bar
def progress_bar(t,len_data,start_time,lap_time):
    t_per = int((t+1)/len_data * 100)
    bar = ("progress:[" + "#"*t_per +" "*(100-t_per)+ "]")
    time.sleep(1e-2)
    elapsed_time = time.time() - start_time
    time_end=int(-(elapsed_time - (elapsed_time*(len_data/(t+1))))/ 60)
    time_per = ((elapsed_time /(elapsed_time*(len_data/(t+1))))*100)
    print("\r {} {}s/{}s end {}m {}%".format(bar,
                                int(elapsed_time),
                                int(elapsed_time*(len_data/(t+1))),
                                time_end,
                                time_per),
                                
          end="")
    lap_time = (time.time() - start_time)/(t+1)
    return lap_time
```
